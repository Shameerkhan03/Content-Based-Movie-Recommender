```python
dataset = movies_mrgd[['id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.duplicated().sum() #checking if there are any duplicates
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
dataset['genres']= dataset['genres'].apply(convert)
```

```python
dataset.head()
```

```python
from collections import Counter
from matplotlib import pyplot as plt

# Flattening the genres list and counting occurrences
all_genres = [genre for sublist in dataset['genres'] for genre in sublist]
genre_counts = Counter(all_genres)

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(genre_counts.keys(), genre_counts.values(), color='skyblue')
plt.xticks(rotation=45)
plt.title('Distribution of Genres')
plt.xlabel('Genres')
plt.ylabel('Frequency')
plt.show()
```

```python
dataset['keywords']= dataset['keywords'].apply(convert)
```

```python
dataset.head()
```

```python
# Flattening keywords and counting
all_keywords = [keyword for sublist in dataset['keywords'] for keyword in sublist]
keyword_counts = Counter(all_keywords).most_common(10)

# Data for plotting
keywords, counts = zip(*keyword_counts)

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(keywords, counts, color='lightgreen')
plt.xticks(rotation=45)
plt.title('Top 10 Most Frequent Keywords')
plt.xlabel('Keywords')
plt.ylabel('Frequency')
plt.show()
```

```python
def convert_cast(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    L.append(i['name'])
    counter += 1
    if counter == 3:
      break
  return L
```

```python
dataset['cast']= dataset['cast'].apply(convert_cast)
```

```python
dataset.head()
```

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
dataset['crew']= dataset['crew'].apply(fetch_director)
```

```python
dataset.head()
```

```python
dataset['overview']=dataset['overview'].apply(lambda x: x.split())
```

```python
dataset.head()
```

```python
dataset['cast'] = dataset['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
dataset['genres'] = dataset['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
dataset['keywords'] = dataset['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
dataset['crew'] = dataset['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
dataset.head()
```

```python
dataset['tags']=dataset['overview']+dataset['genres']+dataset['keywords']+dataset['cast']+dataset['crew']
```

```python
dataset.head()
```

```python
new_df=dataset[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
from wordcloud import WordCloud

# Creating a combined text of all tags
all_tags = ' '.join(new_df['tags'])

# Generating word cloud
wordcloud = WordCloud(width=800, height=400, background_color='black').generate(all_tags)

# Displaying the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Tags')
plt.show()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
features=cv.get_feature_names_out()
```

```python
for i, feature in enumerate(features, start=1):
    print(f"{i}: {feature}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors) #har vector ki sary 4608 vectors k sath similarity calculate hogi bw 0,1
```

```python
similarity
```

```python
import seaborn as sns

# Creating a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(similarity[:20, :20], cmap='viridis', annot=False)
plt.title('Cosine Similarity Heatmap (First 20 Movies)')
plt.xlabel('Movie Index')
plt.ylabel('Movie Index')
plt.show()
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
